```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/train.csv')
df_test = pd.read_csv('Data/test.csv')
```

```python
#원핫인코딩
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output = False).set_output(transform ='pandas')
df_ob = df[[i for i in df.columns if df[i].dtype == 'object']]
encoding = encoder.fit_transform(df_ob)
encoding
```

```python
df = pd.concat([df,encoding], axis = 1)
```

```python
df.drop([i for i in df.columns if df[i].dtype == 'object'], axis = 1, inplace = True)
df.drop(['id'], axis = 1, inplace = True)
df
```

**머신러닝**

```python
#선형회귀
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
df_li = df.copy()
df_li.drop(['exam_score'], axis = 1, inplace = True)
y = df['exam_score']
X_train, X_test, y_train, y_test = train_test_split(df_li, y, test_size=0.2, random_state=42)
# 정규화 (Ridge/Lasso는 정규화된 데이터에서 안정적)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
models ={
    "lnear" : LinearRegression(),
    "Lasso" : Lasso(alpha = 0.1),
    "Ridge" : Ridge(alpha = 0.9)
} 
scores = {}
for name, model in models.items():
    if name == "lnear":
        model.fit(X_train, y_train)  # 선형회귀는 정규화 없이도 가능
        y_pred = model.predict(X_test)
    else:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
    scores[name] = r2_score(y_test, y_pred)
print(scores)
#세게 다 비슷
```

```python
#knn // n=5 : 0.718, n=10 : 0.74, n=100 : 0.754
from sklearn.neighbors import KNeighborsRegressor
```

```python
model = KNeighborsRegressor(n_neighbors=100)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
score = r2_score(y_test, pred)
print("KNN:",score)
```

```python
#의사결정트리
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
dt = dt.fit(X_train_scaled, y_train)
pred = dt.predict(X_test_scaled)
```

```python
score = r2_score(y_test, pred)
print("Dicision_tree",score)
```

```python
#Boosting
from sklearn.ensemble import GradientBoostingRegressor
n_estimators = [50, 100, 200, 300, 400, 500]
for i in n_estimators:
    gb = GradientBoostingRegressor(n_estimators=i, random_state=42)
    gb.fit(X_train_scaled, y_train)
    pred = gb.predict(X_test_scaled)
    score = r2_score(y_test, pred)
    print(f"Gradient_Boosting n_estimators={i} :",score)  
```

```python
#XGBoost
from xgboost import XGBRegressor
n_estimators = [100, 200, 300, 400]
for i in n_estimators:
    xgb = XGBRegressor( n_estimators=i, 
                        random_state=42)
    xgb.fit(X_train_scaled, y_train)
    pred = xgb.predict(X_test_scaled)
    score = r2_score(y_test, pred)
    print(f"XGBoost n_estimators={i} :",score)
```

```python
#LightGBM
from lightgbm import LGBMRegressor
n_estimators = [100, 200, 300, 400]
for i in n_estimators:
    lgbm = LGBMRegressor(n_estimators=i, random_state=42)
    lgbm.fit(X_train_scaled, y_train)
    pred = lgbm.predict(X_test_scaled)
    score = r2_score(y_test, pred)
    print(f"LightGBM n_estimators={i} :",score)
```
